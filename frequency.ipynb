```python
with open('semantic_tokenizer/ekler_v02.json', 'r', encoding='utf-8') as f:
    ekler = json.load(f)

len(ekler)
```

```python
for ek in ekler:
  if ek in kokler:
    # remove from dict
    del kokler[ek]

len(kokler)

# first 20000
kokler_20000 = dict(list(kokler.items())[:20000])

id = 0
for key, value in kokler_20000.items():
  print(f"{id} {key} {value}")
  id += 1
  kokler_20000[key] = id

with open('kokler_20000.json', 'w', encoding='utf-8') as f:
    json.dump(kokler_20000, f, ensure_ascii=False)
```
