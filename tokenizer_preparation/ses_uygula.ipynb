```python
import json

with open('kokler_20000.json', 'r') as f:
    kokler_20000 = json.load(f)

len(kokler_20000)
```

```python
silinecekler = open('semantic_tokenizer/silinecekler.txt', 'r').read().split('\n')
len(silinecekler), silinecekler[:10]
```

```python
with open('semantic_tokenizer/yumusamislar_dict.json', 'r', encoding='utf-8') as f:
  yumusamislar_dict = json.load(f)
```

```python
with open('semantic_tokenizer/yumusamislar_dict.json', 'w', encoding='utf-8') as f:
  json.dump(yumusamislar_dict, f, ensure_ascii=False)
```

```python
silinecek_kokler = []
for silinecek in silinecekler:
  if silinecek in yumusamislar_dict:
    silinecek_kokler.append(silinecek)
    if silinecek in kokler_20000:
      del kokler_20000[silinecek]

with open('semantic_tokenizer/silinecekler_kokler.txt', 'w', encoding='utf-8') as f:
  for kok in silinecek_kokler:
    f.write(f"{kok}\n")

len(kokler_20000)
```

```python
for key, value in yumusamislar_dict.items():
  if key in kokler_20000 and value not in kokler_20000 and key not in silinecekler:
    kokler_20000[value] = kokler_20000[key]
    print(key, value)

len(kokler_20000)
```

```python
with open('semantic_tokenizer/dusmeler.json', 'r', encoding='utf-8') as f:
  dusmeler = json.load(f)

for key, value in dusmeler.items():
  if value in kokler_20000:
    kokler_20000[key] = kokler_20000[value]

len(kokler_20000)
```

```python
edatlar_dict = json.load(open('semantic_tokenizer/edatlar_v01.json', 'r', encoding='utf-8'))

edatlar = list(edatlar_dict.keys())

for i in range(len(edatlar)):
  edatlar_dict[edatlar[i]] = i

with open('semantic_tokenizer/edatlar_v01.json', 'w', encoding='utf-8') as f:
  json.dump(edatlar_dict, f, ensure_ascii=False)
```

```python
ekler_dict = json.load(open('semantic_tokenizer/ekler_v01.json', 'r', encoding='utf-8'))

ekler = list(ekler_dict.keys())

previous_value = ekler_dict[ekler[0]]
counter = 0
for i in range(len(ekler)):
  if ekler_dict[ekler[i]] != previous_value:
    counter += 1
  previous_value = ekler_dict[ekler[i]]

  ekler_dict[ekler[i]] = counter
with open('semantic_tokenizer/ekler_v03.json', 'w', encoding='utf-8') as f:
  json.dump(ekler_dict, f, ensure_ascii=False)
```
